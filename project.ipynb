# Selecting Features

```python
df=df[['Date','TempHighF','TempAvgF','TempLowF','HumidityHighPercent','HumidityAvgPercent','HumidityLowPercent','WindHighMPH','WindAvgMPH','PrecipitationSumInches']]
df.head()
```

# parsing dates

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace = True)
```

```python
df.head()
```

```python
monthly_avg_high = df['TempHighF'].resample('ME').mean()
print(monthly_avg_high)
```

```python
monthly_avg_low = df['TempLowF'].resample('ME').mean()
print(monthly_avg_low)
```

# Visualization

```python
plt.plot(df.index,df['TempHighF'])
plt.title("Date-Temperature graph")
plt.ylabel("Highest Temperature (°F)")
plt.xlabel("Date")
plt.show()
```

```python
plt.plot(df.index,df['TempLowF'],color = 'y')
plt.title("Date-Temperature graph")
plt.ylabel("Lowest Temperature (°F)")
plt.xlabel("Date")
plt.show()
```

```python
plt.plot(monthly_avg_high.index, monthly_avg_high, marker='*', linestyle='--', color='c')
plt.xlabel('Month')
plt.ylabel('Average High Temp (°F)')
plt.title('Monthly Average Highest Temperature')
plt.show()
```

```python
plt.plot(monthly_avg_low.index, monthly_avg_low, marker='*', linestyle='--', color='m')
plt.xlabel('Month')
plt.ylabel('Average Low Temp (°F)')
plt.title('Monthly Average Lowest Temperature')
plt.show()
```
